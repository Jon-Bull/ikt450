# Theory

### 1. **Distribution of Individual Features**

   - **Purpose**: To understand the spread, central tendency, and shape of the data (e.g., normal distribution, skewness, outliers).
   - **Relevant Plots**:
     - **Histogram** (`sns.histplot`, `plt.hist`): Shows the frequency distribution of a single feature.
     - **KDE Plot** (`sns.kdeplot`): Displays the probability density function of a feature.
     - **Box Plot** (`sns.boxplot`): Highlights the quartiles, median, and potential outliers in the data.
     - **Violin Plot** (`sns.violinplot`): Combines box plot and KDE to show distribution and density.

   - **Example**:
     ```python
     sns.histplot(df['Feature1'], bins=20, kde=True)
     plt.title('Distribution of Feature1')
     plt.show()
     ```

### 2. **Relationships Between Features**

   - **Purpose**: To identify correlations, interactions, and dependencies between features.
   - **Relevant Plots**:
     - **Scatter Plot** (`sns.scatterplot`, `plt.scatter`): Shows relationships between two continuous features.
     - **Pair Plot** (`sns.pairplot`): Visualizes pairwise relationships across multiple features.
     - **Joint Plot** (`sns.jointplot`): Displays the relationship between two variables, along with their marginal distributions.
     - **Correlation Heatmap** (`sns.heatmap`): Visualizes the correlation matrix, showing the strength and direction of linear relationships between features.

   - **Example**:
     ```python
     sns.scatterplot(x='Feature1', y='Feature2', data=df, hue='Outcome')
     plt.title('Feature1 vs Feature2')
     plt.show()
     ```

### 3. **Distribution of Categorical Features**


   - **Purpose**: To understand the frequency and distribution of categories, which helps in understanding class imbalance or the diversity of categories.
   - **Relevant Plots**:
     - **Count Plot** (`sns.countplot`): Shows the frequency of each category in a categorical feature.
     - **Bar Plot** (`sns.barplot`): Displays the average value of a continuous feature for each category in a categorical feature.

   - **Example**:
     ```python
     sns.countplot(x='CategoricalFeature', data=df)
     plt.title('Distribution of CategoricalFeature')
     plt.show()
     ```

### 4. **Class Imbalance**


   - **Purpose**: To check for imbalanced classes in the target variable, which can impact model performance.
   - **Relevant Plots**:
     - **Count Plot** (`sns.countplot`): Useful for visualizing the distribution of the target variable.
     - **Pie Chart** (`plt.pie`): Displays the proportion of each class.

   - **Example**:
     ```python
     sns.countplot(x='Outcome', data=df)
     plt.title('Class Distribution in Outcome')
     plt.show()
     ```

### 5. **Outliers Detection**


   - **Purpose**: To identify and potentially treat outliers that might skew the analysis or affect model performance.
   - **Relevant Plots**:
     - **Box Plot** (`sns.boxplot`): Helps in identifying outliers in continuous features.
     - **Violin Plot** (`sns.violinplot`): Combines distribution shape and outlier detection.

   - **Example**:
     ```python
     sns.boxplot(x='Feature1', data=df)
     plt.title('Box Plot of Feature1')
     plt.show()
     ```

### 6. **Feature Correlation and Multicollinearity**


   - **Purpose**: To identify features that are highly correlated, which could lead to multicollinearity and affect model interpretability.
   - **Relevant Plots**:
     - **Correlation Heatmap** (`sns.heatmap`): Visualizes the correlations between features.
     - **Pair Plot** (`sns.pairplot`): Shows pairwise feature relationships.

   - **Example**:
     ```python
     corr_matrix = df.corr()
     sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
     plt.title('Correlation Matrix')
     plt.show()
     ```

### 7. **Missing Values**


   - **Purpose**: To identify missing data patterns, which is crucial for data cleaning and imputation.
   - **Relevant Plots**:
     - **Heatmap for Missing Values** (`sns.heatmap`): Visualizes where the missing data is located.
     - **Bar Plot** (`sns.barplot`): Can show the number of missing values in each feature.

   - **Example**:
     ```python
     sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
     plt.title('Missing Values Heatmap')
     plt.show()
     ```

### 8. **Dimensionality Reduction**


   - **Purpose**: To reduce the number of features while retaining as much information as possible.
   - **Relevant Plots**:
     - **PCA Plot** (Using Matplotlib after applying PCA): Visualizes the first two principal components of the data.
     - **Pair Plot of Principal Components**: After dimensionality reduction, use `pairplot` to visualize relationships in reduced dimensions.

   - **Example**:
     ```python
     from sklearn.decomposition import PCA
     pca = PCA(n_components=2)
     pca_result = pca.fit_transform(df.drop('Outcome', axis=1))
     plt.scatter(pca_result[:, 0], pca_result[:, 1], c=df['Outcome'], cmap='viridis')
     plt.title('PCA of Features')
     plt.show()
     ```

### 9. **Clustering**


   - **Purpose**: To identify potential groupings or clusters in the data, which can inform feature engineering or highlight natural divisions in the data.
   - **Relevant Plots**:
     - **Clustermap** (`sns.clustermap`): Performs hierarchical clustering and visualizes it as a heatmap.
     - **Dendrogram** (Using Scipy): Visualizes the hierarchy of clusters.

   - **Example**:
     ```python
     sns.clustermap(df.drop('Outcome', axis=1), cmap='viridis', standard_scale=1)
     plt.title('Clustermap of Features')
     plt.show()
     ```

### Summary of Visualization Techniques for ML Data Analysis



| **Area of Interest**                | **Relevant Plots**                                                                                       | **Purpose** |
|-------------------------------------|-----------------------------------------------------------------------------------------------------------|-------------|
| **Distribution of Features**        | Histogram, KDE Plot, Box Plot, Violin Plot                                                                | Understand the spread, shape, and presence of outliers |
| **Feature Relationships**           | Scatter Plot, Pair Plot, Joint Plot, Correlation Heatmap                                                   | Identify correlations, interactions, and dependencies |
| **Categorical Feature Distribution**| Count Plot, Bar Plot                                                                                       | Examine the distribution and frequency of categorical features |
| **Class Imbalance**                 | Count Plot, Pie Chart                                                                                      | Check for imbalance in target classes |
| **Outliers Detection**              | Box Plot, Violin Plot                                                                                      | Detect and visualize outliers |
| **Feature Correlation**             | Correlation Heatmap, Pair Plot                                                                             | Identify multicollinearity and correlations |
| **Missing Values**                  | Missing Values Heatmap, Bar Plot                                                                           | Detect and understand missing data patterns |
| **Dimensionality Reduction**        | PCA Plot, Pair Plot of Principal Components                                                                | Visualize data in reduced dimensions |
| **Clustering**                      | Clustermap, Dendrogram | Discover natural groupings in the data |

### Conclusion



This overview maps typical areas of interest in ML data exploration to specific visualizations that are most effective for each task. By selecting the appropriate visualization, you can gain deeper insights into your data, leading to better feature engineering and ultimately improved model performance.

````python
# Function to plot only the diagonal of the pairplot
def plot_diagonal(df, diag_kind='kde'):
    # Create subplots for the diagonal
    num_features = len(df.columns) - 1  # Exclude the target variable if included
    fig, axes = plt.subplots(nrows=1, ncols=num_features, figsize=(num_features * 3, 3))

    for i, feature in enumerate(df.columns[:-1]):  # Exclude the last column (Outcome)
        if diag_kind == 'kde':
            sns.kdeplot(data=df, x=feature, hue='Outcome', ax=axes[i], fill=True)
        elif diag_kind == 'hist':
            sns.histplot(data=df, x=feature, hue='Outcome', ax=axes[i], kde=True)
        axes[i].set_title(feature)

    plt.tight_layout()
    plt.show()

# Call the function to plot the diagonal
plot_diagonal(df)
````
